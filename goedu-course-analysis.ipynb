```python
df2.insert(loc=5,column='Almas', value=[88,89,44,55,66,77,44])
```

```python
sum_row = df2.sum(axis=0).to_frame().T
sum_row.index = ['Total']  # Rename the index to 'Total'
```

```python
df2.tail()
```

```python
sum_row = df2.iloc[:, 1:].sum().to_frame().T
sum_row.insert(0, 'Unnamed: 0', 'Total')

# Append the sum row to the dataframe
df2 = pd.concat([df2, sum_row], ignore_index=True)
df2
```

```python
# Calculate the maximum score for each student
max_row = df2.iloc[:6, 1:6].max().to_frame().T
max_row.insert(0, 'Unnamed: 0', 'Highest')

# Append the maximum row to the dataframe
df2 = pd.concat([df2, max_row], ignore_index=True)
```

```python
df2.tail()
```

```python
max_row = df2.iloc[:6, 1:6].min().to_frame().T
max_row.insert(0, 'Unnamed: 0', 'Minimum')

# Append the maximum row to the dataframe
df2 = pd.concat([df2, max_row], ignore_index=True)
```

```python
df2.tail()
```
